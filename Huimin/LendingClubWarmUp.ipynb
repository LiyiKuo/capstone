```python
#change issue date and last payment date column data types to date
df_raw['issue_d']=pd.to_datetime(df_raw['issue_d'])
df_raw['last_pymnt_d']=pd.to_datetime(df_raw['last_pymnt_d'])
```

```python
#calculate duration of loan
df_raw['duration_of_loan_months'] = ((df_raw.last_pymnt_d - df_raw.issue_d)/np.timedelta64(1, 'M'))
```

```python
#Drop 16 rows. Missing data on the "last payment date"
df_raw=df_raw.dropna(subset=['last_pymnt_d', 'issue_d', "duration_of_loan_months"])
df_raw['duration_of_loan_months'] = df_raw['duration_of_loan_months'].astype(int)
```
